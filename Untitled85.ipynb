```python
cut_map={'Fair':1,"Good":2,'Very Good':3,'Premium':4,'Ideal':5}
df.cut = df.cut.map(cut_map)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
```

```python
p_lr = Pipeline([('scaler1 ' , StandardScaler()) , ('lr' , LinearRegression())])
p_dtt = Pipeline([('scaler2 ' , StandardScaler()) , ('dtt' , DecisionTreeRegressor())])
p_svr = Pipeline([('scaler3 ' , StandardScaler()) , ('svr' , SVR())])
p_rdf = Pipeline([('scaler4' , StandardScaler()) , ('rdf' , RandomForestRegressor())])

pipelines = [p_lr , p_dtt , p_svr , p_rdf]

for pipe in pipelines:
  pipe.fit(x_train , y_train)
  y_pred = pipe.predict(x_test)
  print('r2_score' , pipe , r2_score(y_pred , y_test))
```
