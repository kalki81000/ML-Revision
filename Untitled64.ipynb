# Data Interpolation >> is a process of estimation

# unknow values from knows values
1.linear interpolation
2.cubic interpolation
3.polynomial interpolation**

```python
# linaer interpolation

import numpy as np

x = np.array([1,2,3,4,5])
y = np.array([1,3,5,7,9])

import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
#interpolate the data using>> interpolation

x_new = np.linspace(1,5,100)
y_interp = np.interp(x_new,x,y)
```

```python
y_interp
```

```python
plt.scatter(x_new , y_interp)
```

```python
#cubic interpolation
x =np.array([1,2,3,4,5])
y = np.array([1,8,27,64,125])
```

```python
plt.scatter(x,y)
```

```python
x_new = np.linspace(1,5,100)
y_interp = np.interp(x_new , x,y)
```

```python
y_interp
```

```python
plt.scatter(y_interp , x_new)
```

```python
from scipy.interpolate import interp1d
```

```python
f= interp1d(x , y , kind = 'cubic')
```

```python
x_new = np.linspace(1 , 5 , 100)
y_interp = f(x_new)
```

```python
y_interp
```

```python
plt.scatter(x_new , y_interp)
```

```python
#polynomial interpolation
x = np.array([1,2,3,4,5])
y = np.array([1,4,9,1,25])
```

```python
p = np.polyfit(x, y , 2)
```

```python
x_new = np.linspace(1,5,100)
y_interp = np.polyval(p , x_new)
```

```python
plt.scatter(x , y)
```

```python
plt.scatter(x_new , y_interp)
```
