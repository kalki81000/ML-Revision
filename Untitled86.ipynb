```python
object_columns = df.dtypes[df.dtypes == 'object'].index
object_columns
```

```python
for col in object_columns:
  print(col , '>>>' , df[col].unique())
```

```python
df_dummy =pd.get_dummies(df[object_columns] , drop_first=True , dtype=int)
```

```python
df.drop(object_columns , axis = 1 , inplace=True)
```

```python
df_ = pd.concat([df , df_dummy] , axis = 1)
```

```python
df_.info()
```

```python
df_.columns
```

```python
x = df_.drop('revenue' , axis = 1)
y = df_['revenue']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = 0.30 , random_state = 1)
```

```python
x_train.shape , x_test.shape
```

```python

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 5)
x_train = pca.fit_transform(x_train)
```

```python
x_test = pca.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsRegressor
knn = KNeighborsRegressor()
knn.fit(x_train , y_train)
```

```python
y_pred = knn.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test , y_pred)
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'n_neighbors': [3, 5, 7, 10],
    #'weights': ['uniform', 'distance'],
    #'metric': ['euclidean', 'manhattan', 'minkowski'],
    'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute']
}
```

```python
grid = GridSearchCV(estimator = knn , param_grid = param_grid , cv = 5 , verbose =3)
```

```python
grid.fit(x , y)
```

```python
grid.fit(x_train , y_train)
```

```python
grid.best_params_
```
